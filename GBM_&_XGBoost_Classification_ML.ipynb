```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,precision_score,confusion_matrix
```

## Import Dataset

```python
breat_cancer_data=load_breast_cancer()
breat_cancer_data
```

```python
breast_cancer_data_df=pd.DataFrame(data=breat_cancer_data.data,columns=breat_cancer_data.feature_names,)
breast_cancer_data_df["Target"]=breat_cancer_data.target
breast_cancer_data_df
```

## 3. Data Understanding 

```python
breast_cancer_data_df.isna().sum()
```

```python
breast_cancer_data_df.describe()
```

```python
breast_cancer_data_df.shape
```

## 4.Data Preparation

```python
x=breast_cancer_data_df.drop("Target",axis=1)
```

```python
std_scaler=StandardScaler()
scaled_x=pd.DataFrame(std_scaler.fit_transform(x))
scaled_x
```

```python
y=breast_cancer_data_df.Target
y
```

## 5. Model Building

```python
from sklearn.ensemble import GradientBoostingClassifier
```

# Gradient Boosting

```python
GBM_classifier=GradientBoostingClassifier( n_estimators=50,max_depth=3)
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_x,y,test_size=0.20,random_state=123,shuffle=True,stratify=y)
```

## 6. Model Training

```python
from datetime import datetime
```

```python
start_time=datetime.now()
GBM_classifier.fit(x_train,y_train)
end_time=datetime.now()
print("Time taken to train Algorithmn :",end_time-start_time)
```

## 7. Model Testing / 8. Model Evaluation 

### training

```python
y_train_pred=GBM_classifier.predict(x_train)
y_train_pred
```

```python
print("Confusion Matrix :\n",confusion_matrix(y_train,y_train_pred))
print("Accuracy Score :",accuracy_score(y_train,y_train_pred))
print("Recall Score :",recall_score(y_train,y_train_pred))
print("Precision Score :",precision_score(y_train,y_train_pred))
```

## testing

```python
y_test_pred=GBM_classifier.predict(x_test)
y_test_pred
```

```python
print("Confusion Matrix :\n",confusion_matrix(y_test,y_test_pred))
print("Accuracy Score :",accuracy_score(y_test,y_test_pred))
print("Recall Score :",recall_score(y_test,y_test_pred))
print("Precision Score :",precision_score(y_test,y_test_pred))
```
