## 2. Building UBCF using Euclideon Distance

```python
from sklearn.metrics import pairwise_distances
```

```python
movie_data
```

```python
euclideon_piv_table = pd.pivot_table( data= movie_data,values="rating",index="userId", columns="movie",).fillna(0)
euclideon_piv_table.columns = movie_data["movie"].unique()
euclideon_piv_table.index   = movie_data["userId"].unique()
euclideon_piv_table
```

```python
euclideon_correlation = pd.DataFrame(pairwise_distances(X = euclideon_piv_table, metric='euclidean',))
euclideon_correlation.columns = movie_data["userId"].unique()
euclideon_correlation.index   = movie_data["userId"].unique()
euclideon_correlation
```

### 3. Building UBCF using Cosine Similarity

```python
cosine_correlation = pd.DataFrame(1- pairwise_distances(X = euclideon_piv_table, metric='cosine',))
cosine_correlation.columns = movie_data["userId"].unique()
cosine_correlation.index   = movie_data["userId"].unique()
cosine_correlation
```
