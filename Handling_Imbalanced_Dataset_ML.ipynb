## 1. Import Necessary Libraries

```python
from sklearn import datasets
import pandas as pd
```

## 2. Import Dataset

```python
breast_cancer_data=datasets.load_breast_cancer()
breast_cancer_data
```

```python
print(breast_cancer_data.DESCR)
```

```python
breast_cancer_data_df=pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
breast_cancer_data_df["Target"] =breast_cancer_data.target
breast_cancer_data_df
```

## 3. Data Understanding

```python
breast_cancer_data_df.shape
```

```python
breast_cancer_data_df.isna().sum()
```

```python
breast_cancer_data_df.dtypes
```

## 4. Model Building

```python
X= breast_cancer_data_df.drop(labels="Target",axis=1)
y=breast_cancer_data_df["Target"]
```

```python
X.shape,y.shape
```

```python
y.value_counts()
```

```python
from matplotlib import pyplot as plt
plt.pie(y.value_counts(),labels=["Benign Cancer","Malignant Cancer"],autopct='%1.f%%')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=12,shuffle=True,stratify=y)
```

```python
print("Training Samples : ",X_train.shape,y_train.shape)
print("Test Samples : ",X_test.shape,y_test.shape)
```

```python
plt.pie(y_train.value_counts(),labels=["Benign Cancer","Malignant Cancer"],autopct='%1.f%%')
plt.show()
```

```python
plt.pie(y_test.value_counts(),labels=["Benign Cancer","Malignant Cancer"],autopct='%1.f%%')
plt.show()
```

## 5. Model Training

```python
%%time
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion='gini',max_depth=3,class_weight={0:2,1:1})
dt_model.fit(X_train,y_train)
```

## 6. Model Testing || 7. Model Evaluation

### Training Data

```python
y_pred_train=dt_model.predict(X_train)
y_pred_train
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
confusion_matrix(y_train,y_pred_train)
```

### Testing Data

```python
y_pred_test=dt_model.predict(X_test)
y_pred_test
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
confusion_matrix(y_test,y_pred_test)
```
