```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
from sklearn.feature_selection import SelectFromModel,RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
credit_card_data=pd.read_csv("credit_card_clean.csv")
credit_card_data
```

```python
credit_card_data=credit_card_data.drop(columns="EDUCATION")
```

```python
credit_card_data.dtypes
```

```python
Label_encoder=LabelEncoder()
```

```python
credit_card_data.isna().sum()
```

```python
credit_card_data["SEX"]=Label_encoder.fit_transform(credit_card_data["SEX"])
credit_card_data["MARRIAGE"]=Label_encoder.fit_transform(credit_card_data["MARRIAGE"])
credit_card_data
```

```python
del credit_card_data["ID"]
credit_card_data
```

```python
X=credit_card_data.drop("DEFAULT",axis=1)
y=credit_card_data["DEFAULT"]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=123,shuffle=True,stratify=y)
```

```python
rf_model=RandomForestClassifier(max_depth=5,)
rf_model.fit(X_train,y_train)
y_pred=rf_model.predict(X_test)
print("accuracy_score   :",accuracy_score(y_test,y_pred))
print("precision_score  :",precision_score(y_test,y_pred))
print("recall_score     :",recall_score(y_test,y_pred))
print("confusion_matrix :\n ",confusion_matrix(y_test,y_pred))
```

```python
rf_model.feature_importances_
```

```python
select_from_model=SelectFromModel(estimator=RandomForestClassifier(),max_features=15)
select_from_model.fit(X_train,y_train)
```

```python
select_from_model.get_support()
```

```python
len(select_from_model.get_support())
```

```python
X.columns[select_from_model.get_support()]
```

```python
X_train_pred_SFM = select_from_model.transform(X_train)
X_test_pred_SFM = select_from_model.transform(X_test)
```

```python
def runRandomForest(X_train, X_test, y_train, y_test):
    rf_model=RandomForestClassifier(max_depth=5,)
    rf_model.fit(X_train,y_train)
    y_pred=rf_model.predict(X_test)
    print("accuracy_score   :",accuracy_score(y_test,y_pred))
    print("precision_score  :",precision_score(y_test,y_pred))
    print("recall_score     :",recall_score(y_test,y_pred))
    print("confusion_matrix :\n ",confusion_matrix(y_test,y_pred))
```

```python
%%time
runRandomForest(X_train, X_test, y_train, y_test)
```

```python
%%time
runRandomForest(X_train_pred_SFM, X_test_pred_SFM, y_train, y_test)
```

```python
rfe_model=RFE(estimator=RandomForestClassifier(), n_features_to_select=12)
rfe_model.fit(X_train,y_train)
```

```python
rfe_model.get_support()
```

```python
X.columns[rfe_model.get_support()]
```

```python
X_train_pred_rfe=rfe_model.transform(X_train)
X_test_pred_rfe=rfe_model.transform(X_test)
```

```python
%%time
runRandomForest(X_train, X_test, y_train, y_test )
```

```python
%%time
runRandomForest(X_train_pred_rfe, X_test_pred_rfe, y_train, y_test )
```

```python
rfe_model=RFE(estimator=GradientBoostingClassifier(max_depth=3), n_features_to_select=12)
rfe_model.fit(X_train,y_train)
```

```python
rfe_model.get_support()
```

```python
X.columns[rfe_model.get_support()]
```

```python
X_train_pred_rfe=rfe_model.transform(X_train)
X_test_pred_rfe=rfe_model.transform(X_test)
```

```python
%%time
runRandomForest(X_train, X_test, y_train, y_test )
```

```python
%%time
runRandomForest(X_train_pred_rfe, X_test_pred_rfe, y_train, y_test )
```

```python
for i in range(1,22):
    rfe_model=RFE(estimator=GradientBoostingClassifier(), n_features_to_select=i)
    rfe_model.fit(X_train,y_train)
    X_train_pred_rfe=rfe_model.transform(X_train)
    X_test_pred_rfe=rfe_model.transform(X_test)
    print("selected feature",i)
    runRandomForest(X_train_pred_rfe, X_test_pred_rfe, y_train, y_test)
    print("=================================")
```
