### Let's build BEST ASSOCIATES for UK as it has more transactions

```python
UK_data=online_retail_data[online_retail_data["Country"]=="United Kingdom"]
UK_data
```

```python
UK_data.isna().sum()
```

```python
del UK_data["CustomerID"]
```

```python
UK_data.dropna(inplace=True)
```

```python
UK_pivot_data=pd.pivot_table( data= UK_data, values="Quantity", index="InvoiceNo", columns="Description",aggfunc="sum").fillna(0)
UK_pivot_data
```

```python
UK_pivot_data_onehot=UK_pivot_data.applymap(lambda x:0 if x==0 else 1)
UK_pivot_data_onehot
```

## 4. Data Mining 

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_temsets=apriori(df=UK_pivot_data_onehot,min_support=0.02, use_colnames=True)
freq_temsets
```

```python
best_associates=association_rules(df=freq_temsets,metric='confidence',min_threshold=0.02)
best_associates
```

```python
best_associates.to_csv("best_anticidents_consquents.csv")
```
